```python
# Importing required libraries and methods

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_selector as selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn import linear_model
from sklearn.linear_model import ElasticNet
from sklearn import svm
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import make_pipeline
from sklearn import set_config
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read test file
df_test = pd.read_csv("test.csv", index_col=0)
print(len(df_test))
df_test.head(3)
```

## Checking for missing data.

### I. Train Data

```python
df_train.info()
```

### Count of missing data (Null Values)

```python
# columns with null values
df_train_col_null = df_train.columns[df_train.isnull().any()==True].tolist()
# null values in these columns (For Counting)
df_train[df_train_col_null].isnull().sum()
```

### II. Test Data

```python
df_test.info()
```

### Count of missing data (Null Values)

```python
# columns with null values
df_test_col_null = df_test.columns[df_test.isnull().any()==True].tolist()
# null values in these columns (For Counting)
df_test[df_test_col_null].isnull().sum()
```

#### The columns "Alley", "PoolQC", "Fence", "MiscFeature" are *dropped* from the feature matrices of both Test and Train Data since the amount of missing values in these columns are very high (more than 50%). This is reasonable since imputing lot of missing values does not add any meaningful value to our analysis.

### Dropping the four columns mentioned above

```python
df_train = df_train.drop(["Alley", "PoolQC", "Fence", "MiscFeature"], axis = 1)

df_test = df_test.drop(["Alley", "PoolQC", "Fence", "MiscFeature"], axis = 1)
```

```python
# Remove rows with missing target
X = df_train.dropna(axis=0, subset=['SalePrice'])
# separate target from predictors
y = X.SalePrice
X.drop(['SalePrice'], axis=1, inplace=True)
```

### Separating Numerical and Categorical columns

```python
numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)

numerical_columns = numerical_columns_selector(X)
categorical_columns = categorical_columns_selector(X)
```

## Imputing the Missing Values

I am using the imputer to replace the missing values and the strategy that has been followed for the numerical variable is with the median, and in categorial with the most frequent value.

```python
# For Numerical Variables
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

# For Categorical Variables
imputer_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

### One-Hot Encoding for Categorical Variables and Standarization for Numerical Variables

* I have used OneHotEncoding to represent the categorical variable into a numerical vector (single representation)
* In the pipeline I have used the standard scaler to standardise the numerical data.

```python
ohe = OneHotEncoder(sparse=False,handle_unknown="ignore")
```

### Building Pre-processing Pipeline

```python
numerical_preprocessor = Pipeline([
        ("imp", imputer),
        ("scl", StandardScaler())
    ])
categorical_preprocessor = Pipeline([
    ("imp_cols", imputer_mode),
    ("ohe", ohe)
])

preprocessor = ColumnTransformer([
    ('categorical_preprocessor', categorical_preprocessor, categorical_columns),
    ('numerical_preprocessor', numerical_preprocessor, numerical_columns)])
```

## Splitting the Train Data into Train and Validation Data

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=2022)
```

### Visualizing the Feature matrix (explanatory Variables) after Pre-Processing

```python
pd.DataFrame(preprocessor.fit_transform(X_train, y_train))
```

# Building Model Pipeline

```python
# Creating model pipeline for regression

model_lasso = make_pipeline(preprocessor, linear_model.Lasso(random_state = 2022))

model_linear = make_pipeline(preprocessor, LinearRegression())

model_elastic = make_pipeline(preprocessor, ElasticNet(random_state=2022))

model_svr = make_pipeline(preprocessor, svm.SVR())

model_boosting = make_pipeline(preprocessor, GradientBoostingRegressor(random_state=2022))
```

## Visualizing the Model Pipeline

```python
set_config(display='diagram')
model_lasso
```

## Fitting the Models

```python
model_linear.fit(X_train, y_train)# fit the model
model_lasso.fit(X_train, y_train)# fit the model
model_elastic.fit(X_train, y_train)# fit the model
model_svr.fit(X_train, y_train)# fit the model
model_boosting.fit(X_train, y_train)# fit the model
```

### Handling missing values for Validation and Test Data

```python
X_valid.info()
```

```python
numerical_columns = numerical_columns_selector(X_valid)
categorical_columns = categorical_columns_selector(X_valid)

data_numerical = X_valid[numerical_columns]
data_categorical = X_valid[categorical_columns]

data_numerical = pd.DataFrame(imputer.fit_transform(data_numerical), columns = data_numerical.columns, index=data_numerical.index)

data_categorical = pd.DataFrame(imputer_mode.fit_transform(data_categorical), columns = data_categorical.columns, index=data_categorical.index)


#join the two dataframes back together
X_valid = pd.concat([data_numerical, data_categorical], axis = 1)
```

```python
X_valid.info()
```

```python
df_test.info()
```

```python
numerical_columns = numerical_columns_selector(df_test)
categorical_columns = categorical_columns_selector(df_test)

data_numerical = df_test[numerical_columns]
data_categorical = df_test[categorical_columns]

data_numerical = pd.DataFrame(imputer.fit_transform(data_numerical), columns = data_numerical.columns, index=data_numerical.index)

data_categorical = pd.DataFrame(imputer_mode.fit_transform(data_categorical), columns = data_categorical.columns, index=data_categorical.index)


#join the two dataframes back together
df_test = pd.concat([data_numerical, data_categorical], axis = 1)
```

```python
df_test.info()
```

### Mean Squared Error for the Regression Models

```python
y_pred_linear = model_linear.predict(X_valid)
y_pred_lasso = model_lasso.predict(X_valid)
y_pred_elastic = model_elastic.predict(X_valid)
y_pred_svr = model_svr.predict(X_valid)
y_pred_boosting = model_boosting.predict(X_valid)

print("Mean-Squared Error for Lasso Model is ", mean_squared_error(np.log(y_pred_lasso),np.log(y_valid)))
print("Mean-Squared Error for Elastic Model is ", mean_squared_error(np.log(y_pred_elastic),np.log(y_valid)))
print("Mean-Squared Error for Support Vector Model is ", mean_squared_error(np.log(y_pred_svr),np.log(y_valid)))
print("Mean-Squared Error for Gradient Boosting Model is ", mean_squared_error(np.log(y_pred_boosting),np.log(y_valid)))
```

## Gradient Boosting Model is chosen since its Mean Squared Error is the lowest.

## Final Prediction on the Test Data

```python
y_pred = model_boosting.predict(df_test)
y_pred
```
